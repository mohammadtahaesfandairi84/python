```python
import numpy as np 
import pandas as ps
import seaborn as sns
from sklearn import linear_model
import sklearn.metrics as sm
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = ps.read_csv('co2.csv')
df.describe()
```

```python
sns.countplot(x='out1' , data=df)
```

```python
plt.subplots(figsize=(6,6))
sns.heatmap(df.corr(),annot=True)
```

```python
x = df.drop("out1" , axis=1)
x = x. drop("cylandr", axis=1)
x = x.drop("fuelcomb" ,axis=1)
y = df.out1
```

```python
x
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size=0.2)
```

```python
x_train
```

```python
reg_linear = linear_model.LinearRegression()
```

```python
reg_linear.fit(x_train,y_train)
```

```python
out1_robot = reg_linear.predict(x_test)
```

```python
out1_robot
```

```python
y_test
```

```python
x_test
```

```python
plt.scatter(x_test, y_test, colorizer="blue")
plt.plot(x_test, out1_robot, color="black", linewidth=2)
plt.show()

```
